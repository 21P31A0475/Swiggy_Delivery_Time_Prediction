### Before Cleaning

```python
# Missing values column list
# age,ratings,restaurant_latitude,restaurant_longitude,delivery_latitude,delivery_longitude,weather,traffic,multiple_deliveries,festival,city_type
# pickup_time_minutes,order_time_hour,distance
```

```python
df["age"] = df.groupby(by="rider_id")["age"].transform(lambda x:x.fillna(x.mean()))
```

```python
df["restaurant_latitude"] = df.groupby(by='city_name')["restaurant_latitude"].transform(lambda x:x.fillna(x.mean()))
```

```python
df["restaurant_longitude"] = df.groupby(by='city_name')["restaurant_longitude"].transform(lambda x:x.fillna(x.mean()))
```

```python
df["delivery_latitude"] = df.groupby(by='city_name')["delivery_latitude"].transform(lambda x:x.fillna(x.mean()))
```

```python
df["delivery_longitude"] = df.groupby(by='city_name')["delivery_longitude"].transform(lambda x:x.fillna(x.mean()))
```

```python
df["ratings"] = df.groupby(by="rider_id")["ratings"].transform(lambda x:x.fillna(x.mean()))
```

```python
df['city_type'] = df.groupby('city_name')['city_type'].transform(lambda x: x.fillna(x.mode()[0]))
```

```python
df["weather"] = df['weather'].fillna(df['weather'].mode()[0])
```

```python
df["traffic"] = df['traffic'].fillna(df['traffic'].mode()[0])
```

```python
df["festival"] = df['festival'].fillna(df['festival'].mode()[0])
```

```python
df["multiple_deliveries"] = df['multiple_deliveries'].fillna(df['multiple_deliveries'].median())
```

```python
df["pickup_time_minutes"] = df['pickup_time_minutes'].fillna(df['pickup_time_minutes'].median())
```

```python
df['order_time_hour'] = df.groupby('order_time_of_day')['order_time_hour'].transform(lambda x : x.fillna(x.median()))
```

```python
df['distance'] = df.groupby('city_name')['distance'].transform(lambda x: x.fillna(x.mean()))
```

```python
df.drop(columns=["order_date","rider_id"],axis=1,inplace=True)
```

```python
df.info()
```

```python
nominal_columns = ["weather","type_of_order","type_of_vehicle","city_type","order_time_of_day","city_name"]
ordinal_columns = ["traffic","festival","order_day_of_week"]
numerical_columns = ["age","ratings","restaurant_latitude","restaurant_longitude","delivery_latitude","delivery_longitude","vehicle_condition",
                     "multiple_deliveries","order_day","order_month","is_weekend","pickup_time_minutes","order_time_hour","distance"]
```

### b) Encoding

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder
from sklearn.preprocessing import StandardScaler,PowerTransformer
```

```python
nominal_transformer = Pipeline([("One Hot Encoding",OneHotEncoder(dtype=int,drop='first',sparse_output=False,handle_unknown='ignore'))])
nominal_transformer
```

```python
traffic_cols = ['low','medium','high','jam']
festival_cols = ["no","yes"]
order_day_of_week_cols = ['sunday', 'monday','tuesday', 'wednesday',
                         'thursday','friday','saturday']
```

```python
ordinal_transformer = Pipeline([("Ordinal Encoding",OrdinalEncoder(dtype=int,categories=[traffic_cols,festival_cols,order_day_of_week_cols]))])
ordinal_transformer
```

```python
numerical_tranformer = Pipeline([("Standard Scaler",StandardScaler()),
                                ("Power Transformer",PowerTransformer(method='yeo-johnson'))])
numerical_tranformer
```

```python
from sklearn.compose import ColumnTransformer
```

```python
preprocessing = ColumnTransformer([("nominal",nominal_transformer,nominal_columns),
                                  ("ordinal",ordinal_transformer,ordinal_columns),
                                  ("numerical",numerical_tranformer,numerical_columns)],remainder='drop')
preprocessing
```

### Hyper Parameter Tuning

```python
import optuna
def Objective(trial):
    params = {
        "n_estimators": trial.suggest_int("n_estimators", 200, 500),
        "max_depth": trial.suggest_int("max_depth", 4, 10),
        "learning_rate": trial.suggest_float("learning_rate", 0.03, 0.2),
        "subsample": trial.suggest_float("subsample", 0.7, 1.0),
        "colsample_bytree": trial.suggest_float("colsample_bytree", 0.7, 1.0)}
    model = XGBRegressor(**params, random_state=23, n_jobs=-1)

    kf = KFold(n_splits=5, shuffle=True, random_state=23)
    score = cross_val_score(model, X_tr, y_train, scoring='r2', cv=kf, n_jobs=-1).mean()
    
    return score
    
study = optuna.create_study(directions=["maximize"])
study.optimize(Objective,n_trials=40,show_progress_bar=True)
```
